```python
import pandas as pd
import numpy as np
import os

# ── LOAD DATA ───────────────────────────────────────────
df = pd.read_csv(
    '../data/processed/sales_clean.csv',
    parse_dates=['date']
)

# ── KPI FUNCTION ────────────────────────────────────────
def compute_kpis(df, year=None):
    """
    Compute executive KPIs
    for a given year.
    """

    d = df[df['year'] == year] if year else df

    d_prev = (
        df[df['year'] == year - 1]
        if year else pd.DataFrame()
    )

    # Core Metrics
    total_revenue = d['revenue'].sum()

    total_profit = d['profit'].sum()

    total_cost = d['cost'].sum()

    total_orders = len(d)

    unique_customers = (
        d['customer_id'].nunique()
    )

    avg_order_value = (
        total_revenue / total_orders
    )

    profit_margin = (
        total_profit / total_revenue
    ) * 100

    # ── YOY GROWTH ─────────────────────────────────────
    yoy_growth = None

    if not d_prev.empty:

        prev_rev = d_prev['revenue'].sum()

        yoy_growth = (
            (total_revenue - prev_rev)
            / prev_rev
        ) * 100

    # ── RETURN KPI DICTIONARY ─────────────────────────
    return {

        "Total Revenue (₹)": round(
            total_revenue,
            2
        ),

        "Total Profit (₹)": round(
            total_profit,
            2
        ),

        "Total Cost (₹)": round(
            total_cost,
            2
        ),

        "Profit Margin (₹)": round(
            profit_margin,
            2
        ),

        "Total Orders": total_orders,

        "Unique Customers": unique_customers,

        "Avg Order Value (₹)": round(
            avg_order_value,
            2
        ),

        "YoY Growth (%)": (
            round(yoy_growth, 2)
            if yoy_growth is not None
            else "N/A"
        ),
    }

# ── PRINT KPIs ──────────────────────────────────────────
for year in [2021, 2022, 2023]:

    kpis = compute_kpis(df, year)

    print("\n" + "=" * 40)

    print(f"KPIs for {year}")

    print("=" * 40)

    for k, v in kpis.items():

        print(f"{k:<30} {v}")

# ── SAVE KPI SUMMARY ────────────────────────────────────
os.makedirs(
    '../data/processed',
    exist_ok=True
)

kpi_rows = [

    (yr, k, v)

    for yr in [2021, 2022, 2023]

    for k, v in compute_kpis(df, yr).items()
]

kpi_df = pd.DataFrame(

    kpi_rows,

    columns=[
        'year',
        'metric',
        'value'
    ]
)

# Save CSV
kpi_df.to_csv(

    '../data/processed/kpis.csv',

    index=False
)

print("\n✅ KPIs saved successfully!")
```
